```python
from pso.run import PSO
from pid.example_imp_pid import ExampleOnePID
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
```

```python
# Initialize PID
pid = ExampleOnePID()

# Initialize PSI
pso = PSO(
        W=0.5
        , c1=0.8
        , c2=0.9
        , n_iterations=100
        , n_particles=10
        , target=0.2
        , target_error=0.000001
    )

# Get tuning value
pid_tuning = pso.run(function_optimization=pid)
print("The best value for tuning PID is: %s" % pid_tuning)
```

```python
tf = pid.get_plant_transfer_function()
t, y = signal.step(tf)
plt.plot(t, y)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Step response - Plant Transfer Function')
plt.grid()
```

```python
tf = pid.get_lti_by_pid(pid_tuning[0], pid_tuning[1], pid_tuning[2])
t, y = signal.step(tf)
plt.plot(t, y)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Step response - With PID Tuning')
plt.grid()
```
